```python
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
import numpy as np

# use the searborn style
plt.style.use("seaborn")
# make interactive plots
plt.ion()

# %matplotlib qt
%matplotlib inline
```

```python
mean = np.array([0, 0], dtype=float)
cov = np.array([[1, 0], [0, 1]], dtype=float)
```

```python
rng = np.random.default_rng()

x, y = rng.multivariate_normal(mean, cov, 10000).transpose()
```

```python
nrows, ncols = 1, 1
fig1 = plt.figure(figsize=(10, 10))
gs = gridspec.GridSpec(nrows, ncols, figure=fig1)
ax1 = fig1.add_subplot(gs[0, 0])

ax1.plot(x, y, marker="o", ms=2., ls="", mfc="k")
```
